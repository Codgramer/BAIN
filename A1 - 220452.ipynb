Q1. Suppose you have a jar with 4 jelly beans: 2 are strawberry-flavored, 1 is blueberry-flavored, and 1 is cinnamon-flavored. You draw one jelly bean at random from the jar.

    a. What is the sample space for this experiment?

    b. We define event A as the jelly bean drawn is strawberry-flavored and event B as The jelly bean drawn is not cinnamon-flavored. What are the probabilities of events A and B?

    c. Are events A and B mutually exclusive? Why or why not?


```python
# a).

# Let's define Sample space for the following case -

Sample_space = ['S', 'S', 'B', 'C']

# Here, 'Sample_space' is an array of all possible combination of outcomes when we draw one jelly bean at random.
# 'S' : Strawberry-flavored jelly beans.
# 'B' : Blueberry-flavored jelly beans.
# 'C' : Cinnamon-flavored jelly beans.

# b).

# Let's define the events A and B -

A = ['S']
B = ['S', 'B']

# Event 'A' is getting a strawberry-flavored jelly bean on the draw. So, set 'A' is an array of favorable outcomes for this event.
# Event 'B' is getting a jelly bean except cinnamon-flavored jelly bean on the draw. So, set 'B' is an array of favorable outcomes for this event.

# Now, let us define the function for calculating 'Probability', assuming all the events are equally likely.

def P(S, E):
  favorable_outcomes = 0.0
  for element in E:
   favorable_outcomes += S.count(element)

  return favorable_outcomes/len(S)

print("P(A) : ", P(Sample_space, A))
print("P(B) : ", P(Sample_space, B))

# c).

"""
Events A and B are not mutually exclusive because both the sets are not disjoint.
If we draw a strawberry-flavored jelly bean, both events A and B occur.
In this scenario, the jelly bean is both strawberry-flavored (Event A) and not cinnamon-flavored (Event B).
Therefore, they are not mutually exclusive.

Also, for events to be mutually exclusive: P (A in B) = 0.
But here, P(A in B) = P (A) = 0.5

"""
```

Q3.  Use PreliZ to explore different parameters for the Gaussian distributions. Use the methods `plot_pdf`, `plot_cdf`, and `plot_interactive`. We discussed the probability mass/density functions and the cumulative density function. But there are other ways to represent functions like the percentile point function ppf. Using the `plot_ppf` method of PreliZ, plot the percentile point function for the BetaBinomial and Gaussian distributions. Can you explain how the ppf is related to the cdf and pmf/pdf?

```python
# If not installed earlier, use command :
#!pip install preliz

# Now, import all the required libraries.

import sys
import arviz as az
import preliz as pz
import matplotlib.pyplot as plt
from scipy.stats import norm

az.style.use("arviz-white")

# Since, Gaussian-Distribution has two parameters to set => mu : mean & sigma : standard deviation.
# Let's take some demo values for 'mu' and 'sigma'.

mu = [0.0, 0.2, 1.0, 3.0, -4.0]
sigma = [1.0, 0.5, 1.0, 1.3, 2.0] # 'sigma > 0' must always satisfy.

# For plotting PDF values.
for index in range(0,5):
    pdf_ = pz.Normal(mu[index], sigma[index]).plot_pdf()

pdf_.set_title('Gaussian : PDFs')
pdf_.set_xlabel('X')
pdf_.set_ylabel('Probability Density')
[line.set_linewidth(3.) for line in pdf_.get_lines()[1::2]]
plt.show()

# For plotting CDF values.
for index in range(0,5):
    cdf_ = pz.Normal(mu[index], sigma[index]).plot_cdf()

cdf_.set_title('Gaussian : CDFs')
cdf_.set_xlabel('X')
cdf_.set_ylabel('Cumulative Probability')
[line.set_linewidth(3.) for line in cdf_.get_lines()[1::2]]
plt.show()

# Now, let's try plot_ppf():

# For Gaussian Curve.
for index in range(0,5):
    ppf_ = pz.Normal(mu[index], sigma[index]).plot_ppf()

ppf_.set_title('Gaussian : PPFs')
ppf_.set_xlabel('X')
ppf_.set_ylabel('Percentile Point')
[line.set_linewidth(3.) for line in ppf_.get_lines()[1::2]]
plt.show()

# For Beta-Binomial Curve.

alpha = [1.0, 2.0, 2.5, 3.0, 2.3]
beta = [1.0, 2.5, 3.0, 7.0, 1.0]

for index in range(0,5):
    ppf_ = pz.BetaBinomial(alpha[index], beta[index], n= 10).plot_ppf()

ppf_.set_title('Beta-Binomial : PPFs')
ppf_.set_xlabel('X')
ppf_.set_ylabel('Percentile Point')
[line.set_linewidth(3.) for line in ppf_.get_lines()[1::2]]
plt.show()

#  Here's how the PPF is related to the CDF and PDF/PMF :

"""
While the CDF gives the cumulative probability up to a certain value of random variable (X <= x),
and the PDF/PMF represents the probability density/mass at each value (X = x),
the PPF performs the reverse operation, giving the value associated with a given cumulative probability.
It implies it is the inverse-CDF function that gives out the Random Variable associated with a given CDF.
They are all closely related.

"""
```

```python
# Let's Try Interactive-Plot:

pz.Normal(mu=0.2, sigma=0.5).plot_pdf(legend= "title")
pz.Normal(mu=0.2, sigma=0.5).plot_cdf(legend= None)
pz.Normal(mu=0.2, sigma=0.5).plot_interactive() # You can change the kind= "cdf" to apply it to CDF curve.
```

Q4. Use PreliZ to compute the moments for the SkewNormal distribution for a different combination of parameters. Generate random samples of different sizes,
like 10, 100, and 1,000, and see if you can recover the values of the first two moments (mean and variance) from the samples. Report your obervations.

```python
# Import all the required libraries.

import numpy as np
import preliz as pz

# Since, parameters in SkewNormal Distribution are, scale-parameter: 'sigmas', location-parameter: 'mus', skewness: 'alphas'.
# Let's define these parameters and sample_sizes.

alphas = [0, 1, 2, 3, 4]
mus = [1, 2, -1, 0.5, 3]
sigmas = [1, 0.5, 2, 1, 0.5] # sigma > 0, must always be satisfied.

sample_sizes = [10, 100, 1000]

def recover_sample(mu, sigma, alpha, sample_size):
    recovered_moments= []
    for size_ in sample_size:
        df = pz.SkewNormal(mu, sigma, alpha).rvs(size=size_, random_state = None)
        df_mean = np.mean(df)
        df_variance = np.var(df, ddof=1)
        recovered_moments.append((df_mean, df_variance))
    return recovered_moments

for index in range(0,5):
    true_mean = pz.SkewNormal(mus[index], sigmas[index], alphas[index]).moments('m')
    true_variance = pz.SkewNormal(mus[index], sigmas[index], alphas[index]).moments('v')
    recovered_moments = recover_sample(mus[index], sigmas[index], alphas[index], sample_sizes)

    print(f"Taken Parameters: alpha={alphas[index]}, mu={mu[index]}, sigma={sigma[index]}")
    print(f"Value of True moments: Mean={true_mean}, Variance={true_variance}")
    print("\nSample sizes:\n")
    for i, size in enumerate(sample_sizes):
        sample_mean, sample_variance = recovered_moments[i]
        print(f"Sample size {size}: Recovered Mean={sample_mean}, Recovered Variance={sample_variance}")
    if(index <= 3):
      print("---------------------------")

"""

Observations :

1). The recovered mean and variance from the samples converge to the true mean and variance on increasing the sample size.
2). For smaller sample sizes, the recovered moments are variable to their true value, but as the sample size grows, the recovered moments become more accurate.
3). The accuracy of moment recovery also depends on the parameters such as (mu, sigma, aplha, tau) of the SkewNormal distribution.

"""
```
